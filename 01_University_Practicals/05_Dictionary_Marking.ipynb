**Create a dictionary with student names and marks, then display the highest-scoring student.**

```python
# Creating a dictionary of student names and their marks
student_marks = {
    'Alice': 85,
    'Bob': 92,
    'Charlie': 78,
    'David': 91,
    'Eve': 88
}

# The max() function iterates over the dictionary keys by default.
highest_scoring_student = max(student_marks, key=student_marks.get)

# Get the highest score
highest_score = student_marks[highest_scoring_student]

# Display the highest-scoring student and their score
print(f"The student with the highest score is: {highest_scoring_student}")
print(f"Their score is: {highest_score}")
```
