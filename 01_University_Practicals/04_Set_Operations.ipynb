**Implement a Python script to perform set operations (union, intersection, difference).**

```python
# Here we will do set operations by using user input system

set_1 = input("Enter elements separated by spaces: ")
set_2 = input("Enter elements separated by spaces: ")
user_set1=set(set_1.split())
user_set2=set(set_2.split())
print(" ")
print("----- Basic Operations -----\n")

# 01  Union Operation
union_a = user_set2|user_set1
print("Union_Operation on the two sets\n")
print(union_a,"\n")

# 02  Intersection Operation
set_a = user_set2 & user_set1
print("Intersection_Operation on the two sets\n")
print(set_a,"\n")

# 03  Difference Operation
dif_a = user_set2 - user_set1
print("Difference_Operation on the two sets\n")
print(dif_a,"\n")
```
