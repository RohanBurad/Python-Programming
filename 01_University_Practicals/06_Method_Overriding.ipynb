**Implement method overriding using polymorphism**

```python
class Animal():
    def speak(self):
      return "The animal makes a sound"

class Dog(Animal):
    def speak(self):
      return "Dog woof's"
class Cat(Animal):
    def speak(self):
      return "Cat Meow's!"

def animal_sound_function(animal_obj):
    print(animal_obj.speak())

animal_instance = Animal()
dog_instance = Dog()
cat_instance = Cat()

animal_sound_function(animal_instance)
animal_sound_function(dog_instance)
animal_sound_function(cat_instance)
```
