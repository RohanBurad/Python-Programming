**Create a Python script to print the Fibonacci series using loops.**

```python
def fibbo_series():
  number = int(input("Enter the Number :"))
  a,b = 0,1
  print(f"The Fibonacci Series of {number} is : ")

  for i in range(number):
    print(a, end=' ')
    c = a
    a = b
    b = c+b
fibbo_series()
```
